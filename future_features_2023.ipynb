```python
avg_net = []
for i in range(len(sub_2019)) :
    net_mover_2019 = sub_2019.loc[i]
    net_mover_2020 = sub_2020.loc[i]
    net_mover_2021 = sub_2021.loc[i]
    net_mover_dif = (int(net_mover_2019['Net_Mover']) - int(net_mover_2020['Net_Mover'])) + (int(net_mover_2020['Net_Mover']) - int(net_mover_2021['Net_Mover']))
    avg_net.append(net_mover_dif / 2)
print(avg_net)
```

```python
avg_imm = []
for i in range(len(sub_2019)) :
    immigration_2019 = sub_2019.loc[i]
    immigration_2020 = sub_2020.loc[i]
    immigration_2021 = sub_2021.loc[i]
    # print(immigration_2019)
    # print(immigration_2020)
    immigration_dif = (int(immigration_2019['immigration']) - int(immigration_2020['immigration'])) + (int(immigration_2020['immigration']) - int(immigration_2021['immigration']))
    avg_imm.append(immigration_dif / 2)
print(avg_imm)
```

```python
avg_ft = []
for i in range(len(sub_2019)) :
    fertility_rate_2019 = sub_2019.loc[i]
    fertility_rate_2020 = sub_2020.loc[i]
    fertility_rate_2021 = sub_2021.loc[i]
    fertility_rate_dif = (int(fertility_rate_2019['Total_fertility_rate']) - int(fertility_rate_2020['Total_fertility_rate'])) + (int(fertility_rate_2020['Total_fertility_rate']) - int(fertility_rate_2021['Total_fertility_rate']))
    avg_ft.append(fertility_rate_dif / 2)
print(avg_ft)
```
